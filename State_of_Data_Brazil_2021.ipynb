Assim que eu fiz a leitura do conjunto de dados, o pandas me retornou a mensagem de atenção *DtypeWarning: Columns (47) have mixed types*. Isso significa que o Pandas não conseguiu identificar corretamente o tipo de dados das 47 variáveis em questão. 

O próximo passo será identificar o tipo de dados das variáveis e buscar corrigir este aviso que o Pandas retornou. 

```python
#dataframe de cópia
df_clean = df.copy()
```

```python
columns = list(df_clean.columns)
```

```python
for column in range(len(columns)):
  columns[column].replace(r'\\', '') #alterar uma das opções da pergunta 7
```

```python
#função para separar as perguntas em Dataframes distintos

def separate_df(n_pattern):
  aux_cols = []
  df_aux = pd.DataFrame()
  df_aux["('P0', 'id')"] = df_clean["('P0', 'id')"] #adiciona o ID em todos os dataframes

  for column in columns:

    if re.search(f"^\('\w[{n_pattern}]_", column): #expressão regular simples para identificar as diversas perguntas
      aux_cols.append(column)  
  
  for col in aux_cols:
    df_aux[col] = df_clean[col]
  
  return df_aux
```

Com a divisão das perguntas em dataframes, é possível analisar com mais profundidade cada pergunta.

```python
df_p1 = separate_df(1); df_p2 = separate_df(2); df_p3 = separate_df(3); df_p4 = separate_df(4); df_p5 = separate_df(5); df_p6 = separate_df(6)
df_p7 = separate_df(7); df_p8 = separate_df(8); df_p9 = separate_df(9);
```
